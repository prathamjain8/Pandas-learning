# Skill challenge

```python
# change the index of the tech dataframe to a 3-level index multiindex containing year, month and day on levels 0,1,2 
# respectively. Set the resulting dataframe to the variables tech_df3

tech_df3 = tech.set_index(['year', 'month', 'day']).drop(columns='index')
```

```python
tech_df3.head()
```

```python
# from the tech_df3 dataset, select all the trading days from the year 2019. In addition, add the existing column axis as
# the 4th level of the multiindex. Assign the resulting series to the variable tech_series.

tech_series = tech_df3.loc[(slice(2019), slice(None), slice(None))].stack()
```

```python
tech_series.head()
```

```python
# from the tech_series dataframe, find the average close price and its standard deviation

# mean and stf for close

tech_series.loc[(slice(None), slice(None), slice(None), 'close')].mean()
```

```python
tech_series.loc[(slice(None), slice(None), slice(None), 'close')].std()
```

```python
tech_series.loc[(slice(None), slice(None), slice(None), 'close')].apply({'Average Price': 'mean', 'Standard devaition': 'std'})
```
