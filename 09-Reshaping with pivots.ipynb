# Skill challenge

```python
# starting with our main dataframe(sat), create a pivot table that summarizes the total student enrollment(Student Enrol)
# across all 5 Borough Which Borough have the highest and lowest high school student bodies.

sat.head()
```

```python
sat.pivot_table(values='Student Enrollment', index='Borough', aggfunc='sum').sort_values(by='Student Enrollment', ascending=False)
```

```python
# Modify the pivot table from the step above to also reflect the average student body size(Student Enrollment) across borough
# in the same pivot table

sat.pivot_table(values='Student Enrollment', index='Borough', aggfunc=['sum', 'mean'])
```

```python
# create a pivot table of high schools from the queens borough, containing the SAT Section and Scores as columns and the 
# School Name as the index. Sort the table in descending order by math section scores.

sat[sat.Borough == 'Queens']\
.pivot_table(index='School Name', values='Score', columns='SAT Section').sort_values(by='Math', ascending=False)
```
