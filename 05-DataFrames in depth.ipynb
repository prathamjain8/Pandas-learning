# Skill Challenge

```python
def popular_string(x):
    if x < 220:
        return "relatively unknown"
    elif x < 600:
        return "Kind of popular"
    elif x < 2000:
        return "popular"
    else:
        return "super-popular"
```

```python
players.page_views.apply(popular_string)
```

```python
# Add the output from the step above as a new column to the players dataframe. Name the column popularity

popularity = players.page_views.apply(popular_string)
players['popularity'] = popularity
```

```python
players.head()
```

```python
# How many "Super-popular" players are there

players.popularity.value_counts()
```
