# Skill challenge 2

```python
# Remove all the food items that contain at least one NaN. Do this is a way that modifies the dataframe, i.e. changes the
# stick. How many food items remains after the exclusions?

nutrition.dropna(how="any", axis=0, inplace=True)
```

```python
nutrition.shape
```

```python
nutrition.head()
```

```python
# From the remaining records, isolate those that have between 20 and 40 mg of Vitamin C per 100g serving. Of these foods,
# which one is the least caloric, i.e. has minimum calories.

nutrition[nutrition.vitamin_c_mg.between(20,40)].calories.nsmallest(1)
```

```python
# How many food items in the dataframe have vitamin c levels of between 2 and 3 (inclusive)standard deviation above the mean

m = nutrition.vitamin_c_mg.mean()
m
```

```python
mp2sd = m + nutrition.vitamin_c_mg.std() * 2
mp2sd
```

```python
mp3sd = m + nutrition.vitamin_c_mg.std() * 3
mp3sd
```

```python
print(m, mp2sd, mp3sd)
```

```python
result_set = nutrition[nutrition.vitamin_c_mg.between(mp2sd, mp3sd)]
```

```python
result_set.vitamin_c_mg.describe()
```
